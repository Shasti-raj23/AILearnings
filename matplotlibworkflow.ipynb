#matplotlib example work 

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1 .prepare data set
x= [1,2,3,4]
y=[11,22,33,44]

# 2. setup plots 
fig, ax= plt.subplots(figsize=(10,10))

#3. add data
ax.plot(x,y)

#  4.Customise the data
ax.set(title="simlpe plot", xlabel="x-axis",ylabel="y-axis")

#save & show (you save  the whole figure)
fig.savefig("images/sampe-plot.png")
```

#Making figures with numpy arrays

```python
x = np.linspace(0,10,100)
x
```

```python
# plot the data
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# use satter plot to make a scatter
fig, ax= plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# make plot fro dictonary
skills ={"java":8,"c":7,"c++":2,"python":7}
fig,ax= plt.subplots()
ax.bar(skills.keys(),skills.values() )
ax.set (title="skills",xlabel="courses",ylabel="rate")
```
