## step 0

```python
import pandas as pd
heart_disease= pd.read_csv("../heart-disease.csv")
heart_disease
```

```python
import numpy as np
```

```python
# create X (feature Matrix )
X = heart_disease.drop("target",axis=1)

#create Y (labels)
Y= heart_disease["target"]
```

```python
#2. choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#we will keep the default hyperparameter
clf.get_params()
```

```python
#3. fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test= train_test_split(X,Y,test_size=0.2)
```

```python
clf.fit(X_train,y_train);
```

```python
X_train
```

```python
#make a prediction
y_label = clf.predict(np.array([0,2,3,4]))
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
# 4 Evaluate the model on the training and testing data
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
X_test # the respresent the dataset attribute and Y represent the label of the values
```

```python
X_train
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
X_test.shape
```

```python
accuracy_score(y_test,y_pred)
```

```python
#5. Impove the efficiency of the algorithm
#try different amot of n_estimators
np.random.seed(42)
for i in range (10,100,10):
    print(f'Trying model with {i} estimators')
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model accuracy on test set:{clf.score(X_test,y_test)*100:.2f}%")
    print(f"")
          
```

```python
#6 . save the data set 
import pickle 
pickle.dump(clf,open("randomforesrt_model1.pkl","wb"))
```
