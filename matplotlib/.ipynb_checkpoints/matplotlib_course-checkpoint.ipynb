## Matplotlib example workflow

```python
#prepare data
x=[1,2,3,4]
y=[11,22,33,44]
#setup plot
fig, ax= plt.subplots(figsize=(10,10))
#plot data
ax.plot(x,y)
#custom plot
ax.set(title ="Simple plt", xlabel="x-axis",ylabel="y-axis")
```

```python
butter_prices ={"Almond butter":10,
               "peanut butter":8,
               "Cashew butter": 12}
fig,ax =plt.subplots()
ax.bar(butter_prices.keys(),butter_prices.values())
ax.set(title="Butter Store",ylabel="Prices $")
fig.show();
```

```python
fig,ax =plt.subplots()
#normal representaion of dictnary doesn't work here we have to make it as list
ax.barh(list(butter_prices.keys()),list(butter_prices.values()))
ax.set(title="Butter horizontal",xlabel="price ($)")
```

## OO method 
object orientated method (this from video 95)

* when you want to plot something quickly you have to use pylplot method.
* when you have to plot something more advanced, use the oo(object orientated method)method.

```python
over_50= heart_disease[heart_disease["age"]>50]
```

```python
over_50.shape
```

```python
over_50.plot(kind="scatter",
               x="age",
               y="chol",
            c="target")
#this just using the pyplot method
```

```python
# OO method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y='chol',
            c='target',
            ax=ax);
```

```python
# OO method from strach
fig , ax = plt.subplots(figsize=(10,6))
scatter= ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target']);
ax.set(title="heart disease and cholestral level",
      xlabel="Age",
      ylabel="Cholesterol");
# legend 
ax.legend(*scatter.legend_elements(),title="Target");
# add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle="--");
```

```python
# video no 97
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))
#add the to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

# customise ax1
ax1.set(title="heart disease and cholestrol levels",
       xlabel="age",
       ylabel="cholestrol");
ax1.legend(*scatter.legend_elements(),title="Target");
# add a horizontal line 
ax1.axhline(y=over_50['chol'].mean(),linestyle="--");

# adding data to the ax2 column

scatter = ax2.scatter(x=over_50["age"],
                     y=over_50['thalach'],
                     c=over_50['target']);
#customise the ax
ax2.set(title="heart disease and max Heart rate",
      xlabel="Age",
      ylabel="Max Heart rate")
#adding a legend element
ax2.legend(*scatter.legend_elements(),title="Target");
# adding a horizotal line 
ax2.axhline(y=over_50['thalach'].mean(),linestyle='--');
#heart subtitle
fig.suptitle("heart disease analysis",fontsize=16,fontweight="bold");
```
